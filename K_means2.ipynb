```python
import sklearn
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
df=pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.head()
```

```python
plt.scatter(df["petal length (cm)"],df["sepal length (cm)"])
```

```python
kn=KMeans(n_clusters=3)
y_pred=kn.fit_predict(df[["petal length (cm)", "sepal length (cm)"]])
y_pred
```

```python
df["cluster"]=y_pred
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

plt.scatter(df1["petal length (cm)"], df1["sepal length (cm)"], color="green")
plt.scatter(df2["petal length (cm)"], df2["sepal length (cm)"], color="red")
plt.scatter(df3["petal length (cm)"], df3["sepal length (cm)"], color="blue")
plt.xlabel("petal length (cm)")
plt.ylabel("sepal length (cm)")
plt.legend()
plt.show()
```

```python
val = kn.cluster_centers_
plt.scatter(val[:,0], val[:,1], color="yellow", marker="*", s=100)
plt.show()
print(kn.cluster_centers_)
print(kn.labels_)
print(y_pred)
print(df["cluster"].describe())



```

```python
from sklearn.metrics import classification_report,accuracy_score
cr=classification_report(y_pred,df["cluster"])
print(cr)
acc=accuracy_score(y_pred,df["cluster"])
print(acc)
```
